```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
heart_data = pd.read_csv('heart.csv')
```

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

#LOGISTIC REGRESSION

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data:- ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on testing Data:- ', testing_data_accuracy)
```

#BUILDING PREDICTIVE SYSTEM


```python
input_data = (62,0,0,138,294,1,1,106,0,1.9,1,3,2)
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)
prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0] == 0):
    print('The person does not have Heart Disease')
else:
    print('The person has Heart Disease')
```

```python
import matplotlib.pyplot as plt
X = heart_data.drop(['target', 'predictions'], axis=1, errors='ignore')

predictions = model.predict(X)
heart_data['predictions'] = predictions

count_not_Heartdisease = np.sum(predictions == 0)
count_Heartdisease = np.sum(predictions == 1)

Average_age_with_no_heart_diease = heart_data[heart_data['predictions'] == 0]['age'].mean()
Average_age_disease = heart_data[heart_data['predictions'] == 1]['age'].mean()

total_count = len(predictions)
percent_no_disease = (count_not_Heartdisease / total_count) * 100
percent_disease = (count_Heartdisease / total_count) * 100

labels = [
        f'No Heart Disease\nAvg Age: {Average_age_with_no_heart_diease:.1f}', 
        f'Heart Disease\nAvg Age: {Average_age_disease:.1f}'
]
sizes = [count_not_Heartdisease, count_Heartdisease]
colors = ['green', 'grey']
explode = (0.1, 0)

plt.figure(figsize=(7, 7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('Percentage of People with and without Heart Disease')

plt.figtext(0.5, 0.01, f'No Heart Disease: {percent_no_disease:.2f}% (Avg Age: {Average_age_with_no_heart_diease:.1f})\n'f'Heart Disease: {percent_disease:.2f}% (Avg Age: {Average_age_disease:.1f})', ha="center", fontsize=12, bbox={"facecolor":"lightgrey", "alpha":0.5, "pad":5})

plt.show()
```
